```python
# Seoul Traffic Congestion Analysis by Road Type
# Course-based project code:
# - Python basic syntax
# - pandas / numpy data processing
# - matplotlib visualization
# - least squares fitting using np.linalg.lstsq
#
# Dataset:
# Seoul passenger car travel speed statistics by time period.
# The uploaded CSV has road types:
# 도시고속도로, 주간선도로, 보조간선도로, 기타 도로
#
# Main idea:
# In the same road type, lower travel speed means heavier congestion.
# Therefore, congestion index is calculated separately for each road type.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import glob
import os

from google.colab import files

uploaded = files.upload()
# ============================================================
# 1. Read CSV file
# ============================================================

def read_csv_korean(path):
    """
    Read Korean CSV file by trying several encodings.
    Korean public data often uses utf-8, cp949, or euc-kr.
    """
    encodings = ["utf-8", "cp949", "euc-kr"]

    for enc in encodings:
        try:
            print(f"Trying encoding: {enc}")
            return pd.read_csv(path, encoding=enc)
        except Exception:
            pass

    raise ValueError("CSV file could not be read. Please check the file encoding.")


# In Google Colab, upload your CSV file first.
# Then the code automatically finds the uploaded CSV file in /content.
csv_files = glob.glob("/content/*.csv")

print("Found CSV files:")
for f in csv_files:
    print("-", f)

if len(csv_files) == 0:
    raise FileNotFoundError("No CSV file found. Please upload the traffic CSV file to Colab.")

DATA_PATH = csv_files[0]
raw = read_csv_korean(DATA_PATH)

print("\nRaw data preview:")
print(raw)
print("\nRaw columns:")
print(raw.columns)


# ============================================================
# 2. Convert the original CSV structure into a clean table
# ============================================================

# The uploaded data has a special structure:
# - Row index 1 contains road types.
# - Row index 2 contains the 2025 speed values.
# - Column names contain time periods.
#
# We convert it into this format:
# year | period | road_type | speed

records = []

# Actual 2025 data row
data_row = raw.iloc[2]

for col in raw.columns:
    if col == "시점":
        continue

    # Remove suffix like ".1", ".2", ".3"
    # Example: "오전-07:00~09:00.2" -> "오전-07:00~09:00"
    period = re.sub(r"\.\d+$", "", str(col))

    # Road type is stored in row index 1
    road_type = raw.loc[1, col]

    # Speed value is stored in row index 2
    speed = data_row[col]

    records.append({
        "year": int(data_row["시점"]),
        "period": period,
        "road_type": road_type,
        "speed": float(speed)
    })

df = pd.DataFrame(records)

print("\nCleaned long-format data:")
print(df)


# ============================================================
# 3. Select four road types and real time periods
# ============================================================

# We exclude "소계" because it is subtotal.
# We also exclude "평균-06:00~22:00" from time-period comparison
# because it is an average over a wide time range, not a specific period.

road_types = ["도시고속도로", "주간선도로", "보조간선도로", "기타 도로"]

period_order = [
    "오전-07:00~09:00",
    "낮-12:00~14:00",
    "오후-17:00~19:00"
]

df_road = df[
    (df["road_type"].isin(road_types)) &
    (df["period"].isin(period_order))
].copy()

print("\nSelected data:")
print(df_road)


# ============================================================
# 4. Calculate congestion index separately for each road type
# ============================================================

def calculate_congestion_by_road(group):
    """
    Calculate congestion index within the same road type.

    In the same road type:
    - highest speed  -> congestion index 0
    - lowest speed   -> congestion index 100

    Formula:
    congestion_index = 100 * (speed_max - speed) / (speed_max - speed_min)
    """
    speed_min = group["speed"].min()
    speed_max = group["speed"].max()

    if speed_max == speed_min:
        group["congestion_index"] = 0
    else:
        group["congestion_index"] = 100 * (speed_max - group["speed"]) / (speed_max - speed_min)

    return group


df_road = df_road.groupby("road_type", group_keys=False).apply(calculate_congestion_by_road)

print("\nData with congestion index:")
print(df_road)


# ============================================================
# 5. English labels for graphs
# ============================================================

period_eng = {
    "오전-07:00~09:00": "Morning",
    "낮-12:00~14:00": "Daytime",
    "오후-17:00~19:00": "Evening"
}

road_eng = {
    "도시고속도로": "Urban Expressway",
    "주간선도로": "Main Arterial Road",
    "보조간선도로": "Auxiliary Arterial Road",
    "기타 도로": "Other Roads"
}

df_road["period_eng"] = df_road["period"].map(period_eng)
df_road["road_eng"] = df_road["road_type"].map(road_eng)


# ============================================================
# 6. Graph 1: Travel speed by road type and time period
# ============================================================

pivot_speed = df_road.pivot(
    index="period",
    columns="road_type",
    values="speed"
).reindex(period_order)

pivot_speed.index = [period_eng[p] for p in pivot_speed.index]
pivot_speed.columns = [road_eng[c] for c in pivot_speed.columns]

plt.figure(figsize=(9, 5))

for col in pivot_speed.columns:
    plt.plot(pivot_speed.index, pivot_speed[col], marker="o", label=col)

plt.xlabel("Time Period")
plt.ylabel("Travel Speed")
plt.title("Travel Speed by Road Type and Time Period")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


# ============================================================
# 7. Graph 2: Congestion index by road type and time period
# ============================================================

pivot_congestion = df_road.pivot(
    index="period",
    columns="road_type",
    values="congestion_index"
).reindex(period_order)

pivot_congestion.index = [period_eng[p] for p in pivot_congestion.index]
pivot_congestion.columns = [road_eng[c] for c in pivot_congestion.columns]

plt.figure(figsize=(9, 5))

for col in pivot_congestion.columns:
    plt.plot(pivot_congestion.index, pivot_congestion[col], marker="o", label=col)

plt.xlabel("Time Period")
plt.ylabel("Congestion Index")
plt.title("Congestion Index by Road Type and Time Period")
plt.ylim(-5, 105)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


# ============================================================
# 8. Four separate graphs: travel speed for each road type
# ============================================================

for road in road_types:
    temp = df_road[df_road["road_type"] == road].copy()
    temp["period"] = pd.Categorical(temp["period"], categories=period_order, ordered=True)
    temp = temp.sort_values("period")

    plt.figure(figsize=(7, 4))
    plt.plot(temp["period_eng"], temp["speed"], marker="o")

    plt.xlabel("Time Period")
    plt.ylabel("Travel Speed")
    plt.title("Travel Speed - " + road_eng[road])
    plt.grid(True)
    plt.tight_layout()
    plt.show()


# ============================================================
# 9. Four separate graphs: congestion index for each road type
# ============================================================

for road in road_types:
    temp = df_road[df_road["road_type"] == road].copy()
    temp["period"] = pd.Categorical(temp["period"], categories=period_order, ordered=True)
    temp = temp.sort_values("period")

    plt.figure(figsize=(7, 4))
    plt.plot(temp["period_eng"], temp["congestion_index"], marker="o")

    plt.xlabel("Time Period")
    plt.ylabel("Congestion Index")
    plt.title("Congestion Index - " + road_eng[road])
    plt.ylim(-5, 105)
    plt.grid(True)
    plt.tight_layout()
    plt.show()


# ============================================================
# 10. Least squares fitting by road type
# ============================================================

# Representative hour for each time period:
# Morning 07:00~09:00 -> 8
# Daytime 12:00~14:00 -> 13
# Evening 17:00~19:00 -> 18

hour_map = {
    "오전-07:00~09:00": 8,
    "낮-12:00~14:00": 13,
    "오후-17:00~19:00": 18
}

df_road["hour"] = df_road["period"].map(hour_map)

fitting_results = []

for road in road_types:
    temp = df_road[df_road["road_type"] == road].copy()
    temp = temp.sort_values("hour")

    x = temp["hour"].values.astype(float)
    y = temp["congestion_index"].values.astype(float)

    # Linear fitting model:
    # y = a0 + a1*x
    X = np.vstack([np.ones_like(x), x]).T

    weights, residuals, rank, s = np.linalg.lstsq(X, y, rcond=None)

    y_fit = X @ weights

    mae = np.mean(np.abs(y - y_fit))
    mse = np.mean((y - y_fit) ** 2)

    fitting_results.append({
        "road_type": road,
        "road_type_eng": road_eng[road],
        "a0": weights[0],
        "a1": weights[1],
        "MAE": mae,
        "MSE": mse
    })

    x_dense = np.linspace(x.min(), x.max(), 100)
    X_dense = np.vstack([np.ones_like(x_dense), x_dense]).T
    y_dense = X_dense @ weights

    plt.figure(figsize=(7, 4))
    plt.scatter(x, y, label="Data")
    plt.plot(x_dense, y_dense, label="Least Squares Fitting")

    plt.xlabel("Representative Hour")
    plt.ylabel("Congestion Index")
    plt.title("Fitting Congestion Trend - " + road_eng[road])
    plt.ylim(-5, 105)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()


# ============================================================
# 11. Fitting result table
# ============================================================

fitting_df = pd.DataFrame(fitting_results)

print("\nLeast squares fitting results:")
print(fitting_df)


# ============================================================
# 12. Short interpretation helper
# ============================================================

print("\nInterpretation:")
print("In the same road type, lower travel speed means heavier congestion.")
print("The congestion index was calculated separately for each road type.")
print("Therefore, each road type is compared fairly across Morning, Daytime, and Evening.")
print("Least squares fitting approximates the congestion trend for each road type.")
```
